```python
import util
import plotting
import evaluation

import numpy as np
import matplotlib.pyplot as plt
plt.rc('xtick', labelsize=18)
plt.rc('ytick', labelsize=18)
plt.rc('axes', labelsize=20)
plt.rc('legend', fontsize=18)
from tqdm import tqdm

import tensorflow as tf
```

```python
#Define GPU device where the code will run on
gpu = tf.config.list_physical_devices('GPU')[0]
print(gpu)
tf.config.experimental.set_memory_growth(gpu, True)
gpu = tf.device('GPU:0/')
```

## Import and reshape data 

```python
base_dir = 'H:/Santi/'
data_path = base_dir+'Data/' 

#Inputs
input_data = util.load_h5(data_path, 'Mito')
print('Inputs'+':', input_data.shape)

#Outputs
output_data = util.load_h5(data_path, 'Proc')
print('Outputs:', output_data.shape)
```

## Load model and test data

```python
model_path = base_dir+'Models/ref_f8_c9_b16/'
filename = 'frames_test'

model = util.load_model(model_path+'model.h5')

frames_test = util.load_pkl(model_path, filename)
input_test = input_data[frames_test]
output_test = output_data[frames_test]
```

```python
from glob import glob
```

```python
all_models_dir = glob(base_dir+'Models/'+'*/model.h5')
models = {}
for model_dir in all_models_dir:
    model_id = model_dir.split('\\')[-2]
    models[model_id] = tf.keras.models.load_model(model_dir)
models
```

## Model predictions

```python
with gpu:
  pred_output_test = model.predict(input_test)[:, :, :, 0]
```

```python
nb_examples = 10
item_id = np.sort(np.random.randint(0, len(output_test), nb_examples))
```

```python
fig, axes = plt.subplots(nb_examples, 2, figsize=(8*2, 8*nb_examples))
for i, ax in zip(item_id, axes):
  plotting.plot_merge(input_test[i], output_test[i], title='output_test', frame=i, ax=ax[0])
  plotting.plot_merge(input_test[i], pred_output_test[i], title='pred_output_test', frame=i, ax=ax[1])
fig.subplots_adjust(wspace=0, hspace=0.15)
plt.show()
```

```python
evaluation.detection_match(output_test, np.array([pred_output_test, pred_output_test]), threshold=1)
```

```python
thresholds = np.linspace(0, 1, 50, endpoint=True)
metrics = evaluation.get_metrics(output_test, pred_output_test, thresholds)
```

```python
model_path = base_dir+'Models/ref_model/f8_c9_b256/'
metrics_b256 = util.load_pkl(model_path, 'metrics')
```

```python
colors = ['tab:blue', 'tab:orange', 'tab:green']
fig, ax = plt.subplots(figsize=(5, 5))
for metric_name, color in zip(metrics, colors):
  ax.plot(thresholds, metrics[metric_name], 'o-', color=color, label=metric_name.title())
  ax.plot(thresholds, metrics_b256[metric_name], 'o-', color=color, alpha=0.5)
ax.set(xlabel='Probability threshold', ylabel='Metric')
ax.legend(loc=(0.23, 0.2))
plt.show()
```
